```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/quote/TSLA/history?p=TSLA'
```

```python
tesla_revenue  = requests.get(url).text 
```

```python
soup = BeautifulSoup(tesla_revenue,"html5lib") 
```

```python
table = soup.find_all('table')
```

```python
len(table)
```

```python
print(soup.prettify())
```

```python
%%html
<!DOCTYPE html>
<html>
<head>
<title>Page Title</title>
</head>
<body>
<h3><b </b></h3>
<p> </p>
<h3> </h3>
<p> </p>
<h3> </h3>
<p> </p>
</body>
</html>
>>> soup.find(text="bad")
u'bad'

>>> soup.i
<i>HTML</i>
```

```python
tesla_revenue = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])

for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date =col[0].text
    Open =col[1].text
    high =col[2].text
    low =col[3].text
    close =col[4].text
    adj_close =col[4].text
    volume =col[5].text
    
    tesla_revenue = tesla_revenue.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True)
```

```python
tesla_revenue
```

```python
tesla_revenue.tail()
```

```python
tag_object.parent
```

```python
sibling_1=tag_object.next_sibling
sibling_1
```

```python
for link in soup.find_all('a',href=True):  # in html anchor/link is represented by the tag <a>

    print(link.get('href'))
```

```python
for link in soup.find_all('img'):# in html image is represented by the tag <img>
    print(link)
    print(link.get('src'))
```
